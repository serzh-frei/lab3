## Задание 1. Построение простого графика

Напишите код, который строит график функции $y = x^2$ на интервале `[−10, 10]`. Добавьте заголовок, подписи осей и сетку.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10, 11, 1)
y = x**2

plt.plot(x, y, label='y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции y = x^2')

plt.grid()

plt.legend()
plt.show()
```

+ $y = x^3$ (красная, сплошная линия).

## Задание 7. Трёхмерный график поверхности

Используя `mpl_toolkits.mplot3d`, постройте 3D-график функции $z = \sin(\sqrt{x^2 + y^2})$
на сетке значений $x, y$ в диапазоне `[−5, 5]`.

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)  

Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d') 

surf = ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title('График функции z = sin((x^2 + y^2)^1/2)', fontsize=20, pad=20)
ax.set_xlabel('Ось X', fontsize=12)
ax.set_ylabel('Ось Y', fontsize=12)
ax.set_zlabel('Ось Z', fontsize=12)

plt.tight_layout()
plt.show()
```

## Задание 8. Множественные подграфики (subplots)

Постройте четыре графика в одной фигуре (2 × 2 сетка):

1. Линейный график $y = x$
2. Парабола $y = x^2$
3. Синус $y = \sin(x)$
4. Косинус $y = \cos(x)$

Добавьте заголовки к каждому подграфику.

```python
x = np.linspace(-np.pi, np.pi, 100)  
y1 = x
y2 = x**2
y3 = np.sin(x)  
y4 = np.cos(x)

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes[0, 0].plot(x, y1, color = 'blue')
axes[0, 0].set_title('Линейная функция: y = x')

axes[0, 1].plot(x, y2, color='red')
axes[0, 1].set_title('Парабола: y = x^2')

axes[1, 0].plot(x, y3, color='green')
axes[1, 0].set_title('Синусоида: y = sin(x)')

axes[1, 1].plot(x, y4, color='yellow')
axes[1, 1].set_title('Косинусоида: y = cos(x)')


plt.show()
```

## Индивидуальное задание 3. Задача на вычисление определённого интеграла с помощью Matplotlib

В каждой задаче требуется:
1. **Построить график** подинтегральной функции.
2. **Вычислить площадь под кривой** на заданном отрезке как значение определенного интеграла.
3. **Закрасить область под графиком**, чтобы визуализировать интеграл.
 

Вычислите площадь под кривой:
$$f(x) = \frac{x^2 + 1}{x + 2}$$
на интервале `[0, 3]`.

+ В каждой задаче **используйте функцию** `fill_between()` для закрашивания площади под кривой.
+ Добавьте **подписи осей и заголовок**.
+ При необходимости, отобразите оси X и Y для лучшей визуализации.

```python
from scipy.integrate import quad

def f(x):
    return ((x**2) + 1) / (x + 2)

x = np.linspace(0, 3, 100)
y = f(x)
plt.plot(x, y, label='(x^2 + 1)/(x+2)')
plt.fill_between(x, y, color='red', alpha=0.5)

plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции (x^2 + 1)/(x+2)')

plt.grid()
plt.legend()
plt.show()

integral, error = quad(f, 0, 3)
print('Значение интеграла:', integral)
```

## Индивидуальное задание 4. Задача на построение 3D-графика с помощью Matpotlib

Во всех задачах требуется:

1. **Построить трехмерный график** функции $f(x, y)$ в заданных пределах.
2. **Использовать библиотеку Matplotlib** для визуализации.
3. **Оформить график**: добавить заголовок, подписи осей и цветовую карту (если уместно).

Гиперболический синус

Постройте график:

$$f(x,y) = \sinh(x)+\cosh(y)$$

где

$$\sinh(x) = \frac{e^x - e^{-x}}{2},$$ $$\cosh(x) = \frac{e^x + e^{-x}}{2}$$

на $x,y \in [-2,2]$
+ Используйте `plot_surface()` для плотных поверхностей.
+ Используйте `plot_wireframe()` для каркасных моделей.
+ Настроите цветовую карту через `cmap`.
+ Добавьте подписи осей и заголовки.

```python
def f(x, y):
    return np.sinh(x) + np.cosh(y)

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)

Z = f(X, Y)

fig = plt.figure(figsize=(12, 8))

ax1 = fig.add_subplot(121, projection='3d')
surf = ax1.plot_surface(X, Y, Z, cmap='viridis')
ax1.set_title('Гиперболические функции: поверхность')
ax1.set_xlabel('Ось X')
ax1.set_ylabel('Ось Y')
ax1.set_zlabel('f(x, y)')
fig.colorbar(surf, ax=ax1, shrink=0.5, aspect=5)

ax2 = fig.add_subplot(122, projection='3d')
wire = ax2.plot_wireframe(X, Y, Z, color='blue')
ax2.set_title('Гиперболические функции: каркас')
ax2.set_xlabel('Ось X')
ax2.set_ylabel('Ось Y')
ax2.set_zlabel('f(x, y)')

plt.tight_layout()
plt.show()
```
